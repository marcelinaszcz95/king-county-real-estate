```python
import pandas as pd
import psycopg2
```

```python
import os
from dotenv import load_dotenv
load_dotenv()
```

```python
DATABASE = os.getenv('DATABASE')
USER_DB = os.getenv('USER_DB')
PASSWORD = os.getenv('PASSWORD')
HOST = os.getenv('HOST')
PORT = os.getenv('PORT')
```

```python
conn = psycopg2.connect(
    database=DATABASE,
    user=USER_DB,
    password=PASSWORD,
    host=HOST,
    port=PORT
)
```

```python
cur = conn.cursor()
cur.execute("SELECT * FROM eda.king_county_house_details kd JOIN eda.king_county_house_sales ks ON kd.id = ks.house_id LIMIT 9")
cur.fetchall()
```

```python
query_string = "SELECT * FROM eda.king_county_house_details kd JOIN eda.king_county_house_sales ks ON kd.id = ks.house_id"
df_psycopg = pd.read_sql(query_string, conn)
```

```python
conn.close()
```
